# 1. PrzykЕ‚ad 4

W tym przykЕ‚adzie wykorzystamy API wysokiego poziomu o nazwie Keras zintegrowane z pakietem TensorFlow (https://www.tensorflow.org). Zastosujemy metody optymalizacji na podstawie spadku wzdЕ‚uЕј gradientu.

Na poczД…tku wskaЕјmy pakiety, ktГіre bД™dziemy wykorzystywaД‡ w projekcie

```python
import numpy as np
import tensorflow as tf
```

oraz wygenerujmy przykЕ‚adowe, losowe dane.

```python
# Parametr okreЕ›lajД…cy iloЕ›Д‡ danych zmiennej niezaleЕјnej x
dataset_training_size = 1000
# Prawdziwe wartoЕ›ci parametrГіw
Wt = 0.2
bt = 0.4


# Dodanie szumu do danych
# noise = 0
noise = np.random.rand(dataset_training_size).astype(np.float32)/20

# ZaЕ‚oЕјony model:  y = Wt* x  + bt + noise
# Generacja danych zmiennej niezaleЕјnej x
x_t = np.linspace(0, 1, dataset_training_size).astype(np.float32)

# Tu moЕјna wygenerowaД‡ wartoЕ›ci nierГіwnomiernie
# x_t = np.random.rand(dataset_training_size).astype(np.float32)

x_training = x_t.reshape(dataset_training_size, 1)

# Sortowanie danych, jak nie zostaЕ‚y stworzone liniowo
# x_training = (np.sort(x_t)).reshape(dataset_training_size,1)

W_training = np.array([Wt])

# Wyliczenie prawdziwych wartoЕ›ci modelu y
y_t = (np.dot(x_training, W_training.T)+bt+noise).astype(np.float32)
y_training = np.transpose(np.array([y_t]))
```

Definicja modelu bД™dzie dziedziczyЕ‚a po klasie Model, a nastД™pnie nastД™puje zdefiniowanie parametrГіw i funkcji modelu prostej regresji liniowej.

```python
# 1. Definicja modelu
class SimpleKerasModel(tf.keras.Model):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    self.n = 1
    self.W = self.add_weight(name="W", shape=(self.n, 1), initializer="ones", trainable=True)
    self.b = self.add_weight(name="b", shape=(1,), initializer="ones", trainable=True)

  def call(self, x, **kwargs):
    return self.W * x + self.b
```

Ostatecznie naleЕјy skonfigurowaД‡ proces uczenia, ustawiД‡ algorytm optymalizacyjny, funkcjД™ kosztu oraz wywoЕ‚anie procedury treningu dla okreЕ›lonych parametrГіw procedury

```python
# WywoЕ‚anie modelu
keras_model = SimpleKerasModel()

# Ustawienie hiperparametrГіw
learning_rate = .007
epochs = 100
batch_size=50

# Ustawienie parametrГіw treningu
keras_model.compile(
    # UЕјycie trybu Eager
    run_eagerly=True,
    # wybГіr optymalizatora
    optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate),
    # wybГіr funkcji straty
    loss="mse"
)

# 5. Trening modelu z ustawionymi hiperparametrami
keras_model.fit(x_training, y_training, epochs=epochs, batch_size=batch_size)

print("\nEstimated W: ", keras_model.W.numpy())
print("\nEstimated b: ", keras_model.b.numpy())
```

# Zadania do wykonania

<p><b>Zadanie 1</b></p>
PrzeprowadЕє eksperymenty dla rГіЕјnych wartoЕ›ci batch-size oraz wspГіЕ‚czynnika uczenia. Analizy przeprowadЕє na podstawie funkcji straty w zaleЕјnoЕ›ci od numeru epoki.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf


def make_training_data(seed=42):
    rng = np.random.default_rng(seed)
    sample_count = 1000
    true_w = 0.2
    true_b = 0.4
    noise = rng.random(sample_count).astype(np.float32) / 20
    x_values = np.linspace(0, 1, sample_count).astype(np.float32)
    x_train = x_values.reshape(sample_count, 1)
    y_train = (x_train * true_w + true_b + noise.reshape(-1, 1)).astype(np.float32)
    return x_train, y_train


def build_model():
    return tf.keras.Sequential([
        tf.keras.layers.Input(shape=(1,)),
        tf.keras.layers.Dense(
            1,
            kernel_initializer=tf.keras.initializers.Ones(),
            bias_initializer=tf.keras.initializers.Ones(),
        ),
    ])


def run_experiment(batch_size, learning_rate, epochs=30):
    tf.keras.backend.clear_session()
    tf.random.set_seed(42)
    x_train, y_train = make_training_data()
    model = build_model()
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate), loss="mse")
    history = model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0, shuffle=False)
    losses = history.history["loss"]
    weights, bias = model.layers[0].get_weights()
    return losses, weights.ravel()[0], bias.ravel()[0]


configs = [
    (10, 0.001),
    (10, 0.007),
    (50, 0.007),
    (100, 0.007),
    (50, 0.05),
]

rows = []
for batch_size, learning_rate in configs:
    losses, weight, bias = run_experiment(batch_size, learning_rate)
    label = f"batch={batch_size}, lr={learning_rate}"
    plt.plot(losses, label=label)
    rows.append({
        "batch_size": batch_size,
        "learning_rate": learning_rate,
        "first_loss": losses[0],
        "final_loss": losses[-1],
        "W": weight,
        "b": bias,
    })

plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.show()

results = pd.DataFrame(rows)
print(results.to_string(index=False, float_format=lambda x: f"{x:.6f}"))
```

<p><b>Wnioski do zadania 1</b></p>

NajmniejszД… koЕ„cowД… stratД™ uzyskaЕ‚ wariant `batch_size=50`, `learning_rate=0.05`: `final_loss=0.000226`, przy parametrach `W=0.210` i `b=0.417`. Bardzo podobny wynik daЕ‚ `batch_size=10`, `learning_rate=0.007` (`final_loss=0.000313`, `W=0.231443`, `b=0.403470`).

Przy tym samym `learning_rate=0.007` wiД™kszy `batch_size` pogarszaЕ‚ wynik po 30 epokach: dla `batch_size=10` strata koЕ„cowa wyniosЕ‚a `0.000313`, dla `50` juЕј `0.004838`, a dla `100` `0.007822`. Z kolei zbyt maЕ‚y `learning_rate=0.001` uczyЕ‚ siД™ wolniej (`final_loss=0.006312`). Na podstawie tych wynikГіw najlepszy z testowanych wariantГіw to `batch_size=50` i `learning_rate=0.05`.

Najbardziej stabilne i sensowne wyniki pochodzД… z ustawieЕ„, w ktГіrych bЕ‚Д…d maleje, ale parametry nie eksplodujД…. Dlatego wnioski powinny opieraД‡ siД™ na konkretnych wartoЕ›ciach metryk i przebiegach straty, a nie na ogГіlnym zaЕ‚oЕјeniu, Ејe wiД™kszy learning rate jest zawsze lepszy albo Ејe wiД™kszy batch_size jest zawsze lepszy i tego typu. CzД™sto te parametry trzeba dobraД‡ do konkretnego problemu, wiД™c trudno tutaj pisaД‡ konkretne zalecenia czy konkretne wartoЕ›ci tych parametrГіw.
